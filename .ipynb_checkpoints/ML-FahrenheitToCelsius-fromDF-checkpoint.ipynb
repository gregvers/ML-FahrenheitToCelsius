## FahrenheitToCelsius conversion using Panda dataframe

```python
import tensorflow as tf
import pandas as pd
import numpy as np
import shutil
print(tf.__version__)
```

```python
train_data = [[0,-17.77777778],[32,0],[100,37.77777778],[212,100]]
train_df = pd.DataFrame(train_data,columns=['Fahrenheit','Celsius'])
print(train_df)
valid_data = [[-10,-23.33333333],[10,-12.22222222],[50,10],[80,26.66666667]]
valid_df = pd.DataFrame(valid_data,columns=['Fahrenheit','Celsius'])
print(valid_df)
test_data = [[-15,-26.11111111],[5,-15],[30,-1.111111111],[60,15.55555556]]
test_df = pd.DataFrame(test_data,columns=['Fahrenheit','Celsius'])
print(test_df)
%matplotlib inline
train_df.plot(x='Fahrenheit', y='Celsius')
```

```python
#define model
OUTDIR = './run'
shutil.rmtree(OUTDIR, ignore_errors = True) # start fresh each time

featcol = [ tf.feature_column.numeric_column('Fahrenheit') ]
model = tf.estimator.LinearRegressor(
      feature_columns = featcol, model_dir = OUTDIR)
```

```python
#train
def make_input_fn(df, num_epochs):
  return tf.estimator.inputs.pandas_input_fn(
    x = df,
    y = df['Celsius'],
    batch_size = 128,
    num_epochs = num_epochs,
    shuffle = True,
    queue_capacity = 1000,
    num_threads = 1
  )
model.train(make_input_fn(train_df, num_epochs = 1000000))
```

```python
def print_rmse(model, name, df):
  metrics = model.evaluate(input_fn = make_input_fn(df, 1))
  print('RMSE on {} dataset = {}'.format(name, np.sqrt(metrics['average_loss'])))
print_rmse(model, 'validation', valid_df)
```

```python
#Predict
import itertools
preds_iter = model.predict(input_fn = make_input_fn(test_df, 1))
print([pred['predictions'][0] for pred in list(itertools.islice(preds_iter, 4))])
```
